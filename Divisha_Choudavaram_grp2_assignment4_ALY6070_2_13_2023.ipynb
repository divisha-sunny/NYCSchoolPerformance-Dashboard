```python
#defining a score by combining the diversities. 
# Normalizing the values using min max method to have all the columns in same scale. 
# MinMaxScaler method scales the values in a column between 0 and 1 by subtracting the minimum value in the column from each value and then dividing by the range of values (maximum value minus minimum value).
# Normalize the columns
# Create a diversity score by summing the percentage of each race
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Calculate the diversity score
diversity_score = school[['asian_per', 'black_per', 'hispanic_per', 'white_per']].sum(axis=1)

# Normalize the diversity score between 0 and 1
scaler = MinMaxScaler()
diversity_score_scaled = scaler.fit_transform(diversity_score.values.reshape(-1,1))

# Add the diversity score column to the dataframe
school['diversity_score'] = diversity_score_scaled

```

```python
school['female_per'] = 1 - school['male_per']
```

```python
from sklearn.preprocessing import MinMaxScaler

# Calculate the gender score
socioeconomic_score = school[['frl_percent', 'ell_percent', 'sped_percent']].sum(axis=1)

# Normalize the diversity score between 0 and 1
scaler = MinMaxScaler()
socioeconomic_score_scaled = scaler.fit_transform(socioeconomic_score.values.reshape(-1,1))

# Add the diversity score column to the dataframe
school['socioeconomic_score'] = socioeconomic_score_scaled
```

```python
school_predictors = school[["SchoolName","socioeconomic_score", "male_per", "diversity_score", "total_enrollment", "Dropped Out - % of cohort", "Total Grads - % of cohort", "avg_sat_score", "high_score_percent", "avg_class_size"]]
school_predictors
```

```python
#Creating college score column

weights = [0.05, 0.05, 0.1, 0.1, 0.1, 0.25, 0.25, 0.05, 0.05]

school_predictors["college_score"] = (school_predictors["socioeconomic_score"] * weights[0]) + (school_predictors["male_per"] * weights[1]) + (school_predictors["diversity_score"] * weights[2]) + (school_predictors["total_enrollment"] * weights[3]) + (school_predictors["Dropped Out - % of cohort"] * weights[4]) + (school_predictors["Total Grads - % of cohort"] * weights[5]) + (school_predictors["avg_sat_score"] * weights[6]) + (school_predictors["high_score_percent"] * weights[7]) + (school_predictors["avg_class_size"] * weights[8])
```

```python
school_predictors
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

scaler = MinMaxScaler()
school_predictors_scaled = scaler.fit_transform(school_predictors.drop("SchoolName", axis=1))
school_predictors_scaled = pd.DataFrame(school_predictors_scaled, columns=school_predictors.columns[1:])

school_predictors_scaled["SchoolName"] = school_predictors["SchoolName"]
school_predictors_scaled
```

```python
import pandas as pd

# Create a new column "success" based on the value of "college_score"
school_predictors_scaled["success"] = "No"
school_predictors_scaled.loc[school_predictors_scaled["college_score"] >= 0.5, "success"] = "Yes"

```

```python
school_predictors_scaled = school_predictors_scaled.rename(columns={'Dropped Out - % of cohort': 'Dropout_Rate',
                        'Total Grads - % of cohort': 'Graduation_Rate',
                        'high_score_percent': 'Average_AP_Score',
                        'college_score': 'schoolScore'})
school_predictors_scaled = school_predictors_scaled.drop("SchoolName", axis=1)
school_predictors_scaled
```

```python
from sklearn.model_selection import train_test_split

# Set the predictors (i.e., the independent variables) and the response variable
X = school_predictors_scaled.drop("success", axis=1)
y = school_predictors_scaled["success"]

# Split the data into training and testing sets, with a test size of 0.2 (80-20 split)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# fit the logistic regression model on the train data
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# print the summary of the model
print(log_reg.intercept_)
print(log_reg.coef_)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Load the data into a pandas DataFrame

# Make predictions on the test set
y_pred = log_reg.predict(X_test)

# Evaluate the model's performance
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
import streamlit as st
import pandas as pd

# define a function to predict success based on the input details
def predict_success(socioeconomic_score, male_per, diversity_score, total_enrollment,
                    dropout_rate, graduation_rate, avg_sat_score, avg_ap_score,
                    avg_class_size, school_score):
    input_data = pd.DataFrame(
        {'socioeconomic_score': [socioeconomic_score],
         'male_per': [male_per],
         'diversity_score': [diversity_score],
         'total_enrollment': [total_enrollment],
         'Dropout_Rate': [dropout_rate],
         'Graduation_Rate': [graduation_rate],
         'avg_sat_score': [avg_sat_score],
         'Average_AP_Score': [avg_ap_score],
         'avg_class_size': [avg_class_size],
         'schoolScore': [school_score]
        }
    )
    prediction = log_reg.predict(input_data)
    return "Successful" if prediction >= 0.5 else "Not Successful"

# create the dashboard
st.title("School Success Prediction")

socioeconomic_score = st.number_input("Socioeconomic Score", min_value=0.0, max_value=100.0, value=50.0, step=1.0)
male_per = st.number_input("Percentage of Male Students", min_value=0.0, max_value=100.0, value=50.0, step=1.0)
diversity_score = st.number_input("Diversity Score", min_value=0.0, max_value=100.0, value=50.0, step=1.0)
total_enrollment = st.number_input("Total Enrollment", min_value=0.0, max_value=10000.0, value=5000.0, step=1.0)
dropout_rate = st.number_input("Dropout Rate", min_value=0.0, max_value=100.0, value=50.0, step=1.0)
graduation_rate = st.number_input("Graduation Rate", min_value=0.0, max_value=100.0, value=50.0, step=1.0)
avg_sat_score = st.number_input("Average SAT Score", min_value=0.0, max_value=1600.0, value=800.0, step=1.0)
avg_ap_score = st.number_input("Average AP Score", min_value=0.0, max_value=5.0, value=2.5, step=0.1)
avg_class_size = st.number_input("Average Class Size", min_value=0.0, max_value=50.0, value=25.0, step=1.0)
school_score = st.number_input("School Score", min_value=0.0, max_value=10.0, value=5.0, step=1.0)

if st.button("Predict"):
    result = predict_success(socioeconomic_score, male_per, diversity_score, total_enrollment,
                             dropout_rate, graduation_rate, avg_sat_score, avg_ap_score,
                             avg_class_size, school_score)
    st.write("The school is predicted to be:", result)
```
